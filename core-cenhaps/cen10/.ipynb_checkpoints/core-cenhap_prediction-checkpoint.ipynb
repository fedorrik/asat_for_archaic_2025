```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score
```

# Train model

```python
# Load and normalize input k-mer count data
raw_counts = pd.read_csv('modern_counts.tsv', sep='\t').set_index('id')
norm_counts = raw_counts.div(raw_counts['cen10h1'] + raw_counts['cen10h2'], axis=0)

# select kmers
norm_counts = norm_counts.drop(['cen10h1', 'cen10h2', 'ch1-6-7', 'ch4/11-3', 'ch4/11-2', 'ch1-6-7'], axis=1)

# Load haplotype labels
metadata = pd.read_csv('metadata.tsv', sep='\t').set_index('id')
flank_cenhaps = metadata[['mat_cenhap', 'pat_cenhap']].dropna().map(lambda x: str(int(x)))

# Keep only samples with haplotype assignments
norm_counts = norm_counts.loc[flank_cenhaps.index]
```

```python
sorted(norm_counts.columns)
```

```python
flank_cenhaps.to_csv('flank_cenhaps.tsv', sep='\t')
```

```python
# Encode haplotype targets and split the dataset
hap_classes = [str(i) for i in range(1, 14)]
mlb = MultiLabelBinarizer(classes=hap_classes)
Y = mlb.fit_transform(flank_cenhaps.values)
X = norm_counts.values

X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X, Y, norm_counts.index, test_size=0.2, random_state=42
)
```

```python
# Plot accuracy vs tree depth to detect overfitting
depths = list(range(1, 10))
train_acc = []
test_acc = []

for depth in depths:
    clf = MultiOutputClassifier(RandomForestClassifier(max_depth=depth, n_estimators=100, random_state=42))
    clf.fit(X_train, y_train)
    train_acc.append(accuracy_score(y_train, clf.predict(X_train)))
    test_acc.append(accuracy_score(y_test, clf.predict(X_test)))

plt.plot(depths, train_acc, label='Train Accuracy')
plt.plot(depths, test_acc, label='Test Accuracy')
plt.xlabel('Tree Depth')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Tree Depth')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Train final model with chosen depth and evaluate accuracy
model = MultiOutputClassifier(RandomForestClassifier(max_depth=6, n_estimators=100, random_state=42))
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print(f"Train accuracy: {train_acc:.4f}")
print(f"Test accuracy: {test_acc:.4f}")
```

```python
# Compute and display feature importance table
importances = np.mean([est.feature_importances_ for est in model.estimators_], axis=0)
feature_importance = pd.DataFrame({
    'kmer': norm_counts.columns,
    'importance': importances
}).sort_values(by='importance', ascending=False)
feature_importance
```

```python
# Convert predictions to haplotype labels and assemble result table
true_train_labels = mlb.inverse_transform(y_train)
pred_train_labels = mlb.inverse_transform(y_train_pred)
true_test_labels = mlb.inverse_transform(y_test)
pred_test_labels = mlb.inverse_transform(y_test_pred)

results = pd.DataFrame({
    'id': list(idx_train) + list(idx_test),
    'split': ['train'] * len(idx_train) + ['test'] * len(idx_test),
    'true': [','.join(sorted(p)) for p in true_train_labels] + [','.join(sorted(p)) for p in true_test_labels],
    'pred': [','.join(sorted(p)) for p in pred_train_labels] + [','.join(sorted(p)) for p in pred_test_labels],
})
results['correct'] = results['true'] == results['pred']
results.set_index('id', inplace=True)
#results.to_csv('results', sep='\t')
results.head()
```

```python
# Check errors
errors = results[results['correct'] == False]
cnt = Counter(errors['split'])
print(cnt)
```

# Predict samples

```python
def predict_core_cenhap(path_to_counts_file, output_suffix='_core-cenhaps.tsv'):
    """
    Predict cenhap haplotypes from k-mer counts using trained model.
    
    Parameters:
        path_to_counts_file (str): Path to input tsv file with k-mer counts (must include cen10h1 and cen10h2).
        output_suffix (str): Suffix for output file name.
        
    Returns:
        pd.DataFrame: Input data with added 'core_cenhaps' column (string label of predicted haplotypes).
    """
    # Load counts
    df = pd.read_csv(path_to_counts_file, sep='\\t').set_index('id')

    # Normalize counts by cen10h1 + cen10h2
    df_norm = df.div(df['cen10h1'] + df['cen10h2'], axis=0)

    # Predict using the trained model (assumes `model` and `mlb` exist)
    X_new = df_norm[norm_counts.columns].values  # use same k-mer columns as training
    y_pred = model.predict(X_new)
    
    # Convert to string labels
    pred_labels = mlb.inverse_transform(y_pred)
    df['core_cenhap'] = [','.join(sorted(haps)) for haps in pred_labels]

    # Save to file
    #out_path = path_to_counts_file.replace('.tsv', output_suffix)
    #df.to_csv(out_path, sep='\t')

    return df[['core_cenhap']]
```
